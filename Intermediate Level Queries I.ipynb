## Query 1

Using the occupations and trainers table, write out a query that returns the average salaries for each job. The actual name of the job should be returned, not the job_id number. Use aliases for all column headers. Order the results by the average salary in descending order. Round the average salary to the nearest integer value.

```python
%%sql
SELECT o.job_title "Job Title", ROUND(AVG(salary)) "Average Salary"
FROM trainers t
JOIN occupations o
ON o.job_id = t.job_id
GROUP BY o.job_title
ORDER BY "Average Salary" DESC
```

 ## Query 2
Using the jobs table which is in the HR schema and the emps table, create a case statement that categorizes each employee’s salary along with returning the actual name of the job each employee has. Use aliases to rename column headers. Order the results by the employee's salary in descending order.
 

```python
%sql SELECT * FROM emps WHERE rownum < 3
```

```python
%sql SELECT * FROM hr.jobs WHERE rownum < 3 
```

```python
%%sql
SELECT last_name|| ', '||first_name "Employee Name", job_title "Job Title",
    CASE  
    WHEN salary BETWEEN 18000 AND 24000 THEN 'Super-high earner'
    WHEN salary BETWEEN 15000 AND 17999 THEN 'High earner'
    WHEN salary BETWEEN 10000 AND 14999 THEN 'Middle-high earner'
    WHEN salary BETWEEN 7500 AND 9999 THEN 'Middle-level earner'
    WHEN salary BETWEEN 5000 AND 7499 THEN 'Lower-middle earner'
    WHEN salary < 5000 THEN 'Low earner'
    END AS "Salary Category"
FROM emps e
JOIN hr.jobs j
ON e.job_id = j.job_id
ORDER BY salary DESC
```

## Query 3
Using the emps table, return the following columns: first_name and last_name as one column, 
email, phone number, and another column which would be their password. The password column needs to contain the first three letters of their first name in all caps and the 4th from last to the second to last letter of their last name with the first letter capitalized. Example: The password Nancy Greenberg would be NANBer. Use aliases for more descriptive column names. Limit the number of rows returned to 15.

## Query 4
Assume that you do not know the department name of where David Austin, Luis Popp, or Nancy Greenburg from the returned records listed above work. Create a view that returns all employees that work in the same department as either David, Luis, or Nancy. Columns to be returned should be the department_id, employee's first and last name, hire_date, salary, commission_pct, and the name of the department. Determine what departments these employees work in and create the view with an appropriate name to match the departments or department ID numbers. 

```python
%%sql
CREATE VIEW IT_FINANCE_EMPS
AS
SELECT e.department_id, department_name, first_name, last_name, hire_date, salary, commission_pct
FROM emps e
JOIN depts d
ON e.department_id = d.department_id
WHERE e.department_id IN(SELECT department_id
                      FROM emps
                      WHERE last_name IN('Austin', 'Popp', 'Greenburg'))
```

## Query 5
Suppose we want to do some analysis on employees who earn commission. Create a view that consists of employees where the commission percent is not null. Aliases will be used to rename columns for the view. Columns to be included in the view are:
* employee_id as "Employee ID"
* first_name as "First Name"
* last_name as "Last Name
* hire_date as "Date Hired"
* salary as "Salary"
* commission_pct as "Commission Percent"
* job_title as "Job Title"
* department_name as "Department Name"

The actual name of the job and department should be included in the view not the ID numbers. Include records where an employee may not be assigned a job title and or to a department. 

```python
%%sql
CREATE VIEW SALES_EMPLOYEES
("Employee ID", "First Name", "Last Name", "Date Hired", "Salary", "Commission Percent", "Job Title", "Department Name")
AS
SELECT employee_id, first_name, last_name, hire_date, salary, commission_pct, job_title, department_name
FROM emps e
FULL JOIN hr.jobs j 
ON e.job_id = j.job_id
FULL JOIN depts d
ON e.department_id = d.department_id
WHERE commission_pct IS NOT NULL
```

## Query 6
One may assume that someone with the word 'manager' in their job title would typically make more than someone without such a job title. Write two queries that compares the average, maximum, minimum, standard deviation, and variance for the salary and commission rate (one query for salary, and one query for commission rate) for sales representatives and sales managers. Round the standard deviation and variance values to two decimal places.

```python
%%sql
SELECT "Job Title", AVG("Salary") "Average Salary", MAX("Salary") "Maximum Salary", MIN("Salary") "Minimum Salary",
ROUND(STDDEV_POP("Salary"),2) "Salary Standard Deviation", ROUND(VARIANCE("Salary"),2) "Salary Variance"
FROM sales_employees
GROUP BY "Job Title"
```

```python
%%sql 
SELECT "Job Title", AVG("Commission Percent") "Average Commission Percent", 
MAX("Commission Percent") "Maximum Commission Percent", MIN("Commission Percent") "Minimum Commission Percent",
ROUND(STDDEV_POP("Commission Percent"),2) "Commission Percent STDDEV", 
ROUND(VARIANCE("Commission Percent"),2) "Commission Percent Variance"
FROM sales_employees
GROUP BY "Job Title"
```

 Based on this data set, sales managers have higher values across the board for both salary and commission percent values compared to sales representatives.

## Query 7
Using a subquery, find the total number of sales representatives who make greater than the sales manager with the minimum salary. 

```python
%%sql
SELECT COUNT("Employee ID") "Sales Representative Count"
FROM sales_employees
WHERE "Salary" > (SELECT MIN("Salary")
                 FROM sales_employees
                 WHERE "Job Title" = 'Sales Manager')
AND "Job Title" = 'Sales Representative'
```


## Query 8
Assuming each employee has never gotten a raise in salary and or commission rate in the time they have worked at their job, write a query that returns the employee's first name, last name, salary, combined salary (salary plus their commission), the total months they have been employed, and their cumulative salary. Round the months employed to two decimal places, and the cumulative salary to two decimal places. When finding the cumulative salary, only round at the very end.
 
 

```python
%%sql
SELECT "First Name", "Last Name", "Salary", "Salary" + ("Salary" * "Commission Percent") "Combined Monthly Salary",
ROUND((SYSDATE - "Date Hired") / 30, 2) "Total Months Employed",
ROUND((SYSDATE-"Date Hired")/30 * ("Salary" + ("Salary" * "Commission Percent")),2) "Cumulative Salary"
FROM sales_employees
```

## Query 9
Suppose you are tasked with finding all employees who were hired after Peter Hall and have a salary higher than Peter. Assume that you do not know when he was hired or what his salary is. Using subqueries, find all employees who were hired after and make a greater salary than Peter Hall. Order the results by the hire date and then order by salary in descending order. 

```python
%%sql
SELECT employee_id, first_name, last_name, hire_date, salary
FROM emps
WHERE hire_date > (SELECT hire_date
                  FROM emps
                  WHERE first_name = 'Peter' AND last_name = 'Hall')
AND salary > (SELECT salary
             FROM emps
             WHERE first_name = 'Peter' AND last_name = 'Hall')
ORDER BY hire_date, salary DESC
```
